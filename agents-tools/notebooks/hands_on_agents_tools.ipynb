### Load the generative LLM model

```python
from langchain_ollama.llms import OllamaLLM
llm = OllamaLLM(model="qwen:32b")
```

### Example of offline hallucination

```python
llm.invoke("Quien es Jokin Aperribay?")
```

### Load tools: internet search and Python interpreter

```python
from langchain.tools import Tool
from langchain.tools import DuckDuckGoSearchRun
from langchain_experimental.utilities import PythonREPL


search = DuckDuckGoSearchRun()

python_repl = PythonREPL()

tools = [
Tool(
    name='DuckDuckGo Search',
    func= search,
    description="A tool to retrieve precise, real-time information from the web based on specific user queries."
),

Tool(
    name = "python repl",
    func=python_repl.run,
    description="A tool when you need to use python to answer a question. You should input python code, you can use it as a calculator."
)
]

print(python_repl.run("print(17*2)"))
print(search.run("Maradona wikipedia"))
```

### Initialize agent with tools

```python
from langchain.agents import initialize_agent

zero_shot_agent = initialize_agent(
    agent="zero-shot-react-description",
    tools=tools,
    llm=llm,
    verbose=True,
    max_iterations=5,
    handle_parsing_errors=True
)
```

### Examples

```python
zero_shot_agent.run(""""Quien es Jokin Aperribay?""")
```

```python
from langchain.globals import set_debug, set_verbose
set_debug(True)
set_verbose(True)
zero_shot_agent.run("""Cuantos es la raiz cuadrada de los años que lleva Jokin Aperribay como presidente de la Real?""")
```
